(5) The club decides to sell the leftover items at half price and everything quickly sells out!  Determine how much money was collected from the clearance sale.  

```python
price_leftover_array = price_array * 0.5
collected_leftover_array = price_leftover_array * quantity_leftover_array

clearance_sale = np.sum(collected_leftover_array)

print(f'The clearance sale brought in ${clearance_sale}')
```

```python
# Easiest route... but less elegant (?)... np.sum(quantity_sold_array * price_array) + np.sum(collected_leftover_array)
total_sales = bake_sale + clearance_sale

print(f'The bake sale brought in ${bake_sale}')
print(f'The clearance sale brought in ${clearance_sale}')
print(f'Total sales were ${total_sales}')
```
