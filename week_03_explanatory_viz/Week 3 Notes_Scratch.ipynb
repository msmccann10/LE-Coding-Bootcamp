#Boxplots
A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed. 

```python
cancer_df = pd.read_csv(cancer_filepath)
```

```python
display(cancer_df.head())
cancer_df.info()
```

```python
# boxplots using matplotlib == plt.boxplot
# format == plt.boxplot([values1, values2], lables=['label1', 'label2'])

malignant = cancer_df.loc[cancer_df['diagnosis']=='M','area_mean'].values
benign = cancer_df.loc[cancer_df['diagnosis']=='B','area_mean'].values
plt.boxplot([malignant,benign], labels=['M', 'B']);
```

```python
# boxplots using pandas == df.boxplot
# format == df.boxplot(column = 'values', by = 'split');

cancer_df.boxplot(column = 'area_mean', by = 'diagnosis');
plt.title('');
plt.suptitle('');
```

```python
# boxplots using seaborn == sns.boxplot()
# format == sns.boxplot(x='split', y='column', data=df)

sns.boxplot(x='diagnosis', y='area_mean', data=cancer_df)
```

```python
# Notched boxplots (allow us to evaluate confidence intervals)

malignant = cancer_df.loc[cancer_df['diagnosis']=='M','area_mean']
benign = cancer_df.loc[cancer_df['diagnosis']=='B','area_mean']
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.boxplot([malignant,benign], notch = True, labels=['M', 'B']);
```

```python
# Make it pretty....

malignant = cancer_df.loc[cancer_df['diagnosis']=='M','area_mean']
benign = cancer_df.loc[cancer_df['diagnosis']=='B','area_mean']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8,5))
boxplots = axes.boxplot([malignant,benign],
           notch = True,
           labels=['M', 'B'],
           widths = .7,
           patch_artist=True,
           medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
           boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4)
          );
# The more you understand any library, the more you can do
boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('red')
plt.xlabel('diagnosis', fontsize = 20);
plt.ylabel('area_mean', fontsize = 20);
plt.xticks(fontsize = 16);
plt.yticks(fontsize = 16);
```

# Correlation and Heatmaps
- If two variables change in the same direction, they have a positive correlation. 
- If two variable change in the opposite direction, they have a negative correlation. 
- We can also quantify the relationship with a numerical value called the "correlation coefficient."  This value measures how strong the relationship is and can range from -1 to 1.  
- If two variables are not related at all, they will have a correlation coefficient of 0! A value of 1 is the strongest positive relationship. A value of -1 is the strongest negative relationship.


---


In general a correlation coefficient greater than 0.7 is considered "strong."

Between 0.5 and 0.7 is considered "moderate".  

Between 0.3 and 0.5 is considered  a "low" correlation.

# Subplots

```python
linearW_df = pd.read_csv(linearW_filepath)
linearW_df.head()

# predicted value is what the model predicted vs the actual value.
```

```python
interceptFilter = linearW_df['intercept'] == True
df_intercept = linearW_df.loc[interceptFilter, :]
df_nointercept = linearW_df.loc[~interceptFilter, :]
```

## MATLAB Style

```python
plt.figure(figsize=(8,4))
# Subplot 1 
plt.subplot(1, 2, 1);
plt.plot(df_intercept['feature'].values, df_intercept['predicted'].values, c = 'r');
plt.scatter(df_intercept['feature'].values, df_intercept['actual'].values, c= 'k');
plt.title('intercept', fontsize = 12);
# Subplot 2
plt.subplot(1, 2, 2);
plt.plot(df_nointercept['feature'].values, df_nointercept['predicted'].values, c = 'r');
plt.scatter(df_nointercept['feature'].values, df_nointercept['actual'].values, c= 'k');
plt.title('no intercept', fontsize = 12);
```

## OOP Style

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(8,4));
# Subplot 1 
axes[0].plot(df_intercept['feature'].values, df_intercept['predicted'].values, c = 'r');
axes[0].scatter(df_intercept['feature'].values, df_intercept['actual'].values, c= 'k');
axes[0].set_title('intercept', fontsize = 12);
# Subplot 2
axes[1].plot(df_nointercept['feature'].values, df_nointercept['predicted'].values, c = 'r');
axes[1].scatter(df_nointercept['feature'].values, df_nointercept['actual'].values, c= 'k');
axes[1].set_title('no intercept', fontsize = 12);
```
