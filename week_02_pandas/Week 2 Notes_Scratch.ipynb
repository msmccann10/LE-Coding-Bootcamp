# Renaming and Deleting Columns


```python
# We will be using the mortgage df again. 
mortgage_df.head()
```

```python
# Renaming Method 1: Dictionary Substitution
mortgage_df = mortgage_df.rename(columns={'Starting Balance': 'starting_balance',
                                          'Interest Paid': 'interest_paid',
                                          'Principal Paid': 'principal_paid'})

mortgage_df.head()
```

```python
# Renaming Method 2: List replacement

mortgage_df.columns = ['month',
                       'starting_balance',
                       'repayment',
                       'interest_paid',
                       'principal_paid',
                       'new_balance',
                       'mortgage_name',
                       'interest_rate']

mortgage_df.head()
```

```python
# Deletion Method 1: Drop Method

mortgage_df = mortgage_df.drop(columns=['new_balance', 'month'])
mortgage_df.head()
```

```python
# Deletion Method 2: Del

del mortgage_df['starting_balance']
mortgage_df.head()
```

#Identifying Missing Data

```python
linear_df = pd.read_csv(linear_filepath) 

display(linear_df.head())
linear_df.info()
```

Missing values in Pandas usually 'NaN' or 'None' 

```python
# Create a pandas string to see where there are/aren't missing values
linear_df['y'].isna().head(10)
```

```python
# Create a filter and apply to the dataframe to find NA values
y_missing = linear_df['y'].isna()

linear_df.loc[y_missing,:]
```

```python
# the ~ can be used to negate (read do the opposite) 

linear_df.loc[~y_missing,:]
```

# Removing or Filling In Missing Data

Sticking with the linear dataset. 

```python
# loading in the data so it is fresh/untampered with....
linear_df = pd.read_csv(linear_filepath) 

display(linear_df.head())
linear_df.info()
```

```python
# option within dropna to do how = 'any' or how = 'all'. default is any.

linear_df.loc[0:10,:].dropna(how = 'any')

# linear_df.loc[0:10,:].dropna() # how not needed because any was default
```

```python
linear_df.loc[0:10, 'y'].fillna(0) # replace with a specific value
```

```python
linear_df.loc[0:10, 'y'].fillna(method = 'bfill') # back fill
```

```python
linear_df.loc[0:10, 'y'].fillna(method = 'ffill') # forward fill
```

```python
linear_df.loc[0:10, 'y'].interpolate(method = 'linear')
```
