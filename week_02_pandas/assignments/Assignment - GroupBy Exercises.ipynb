This data comes from the main door of the CalIt2 building at UCI. Observations come from 2 data streams (people flow in and out of the building), over 15 weeks, 48 time slices per day (half hour count aggregates).

```python
# Load Data: 
file = "https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data"
df = pd.read_csv(file, header = None)

display(df.head())
df.info()
```

```python
irvine_df = df.rename(columns = {0: "Flow_ID",
                          1: "Date",
                          2: "Time",
                          3: "Count"})

irvine_df.head()
```

#Selecting Data
1. Select all data for the date July 24 2005 having flow id=7.
2. From the data selected in step 1, sort the result on Count in descending order and output the top 10 rows. Assign the output to a variable named top10.

```python
# create filters
flow_filter = irvine_df['Flow_ID'] == 7
date_filter = irvine_df['Date'] == "07/24/05"
selected_df = irvine_df.loc[flow_filter & date_filter, :]
```

```python
# Get the top 10
top10 = selected_df.sort_values(by = ['Count'], ascending=False).head(10)
top10
```

#Apply function
1. For the 10 rows outputted above, use Pandas Apply function to subtract lowest value of the 10 from all of them and then output the average value of the resulting counts

```python
#top10['Count'] - top10['Count'].min()
# test_df = top10
# test_df['Count'] = test_df['Count'].apply(lambda x: x- test_df['Count'].min())

sub = top10['Count'].apply(lambda x: x - top10['Count'].min())
mean = sub.mean()

display(sub)
print("mean is: ", mean)

# all in one function/line
# top10['Count'].apply(lambda x: x - top10['Count'].min()).mean()
```

#Grouping
1. Select data in the month of August 2005 having flow id=7
2. Group the data based on date and get the max count per date

```python
# filters
flow_filter = irvine_df['Flow_ID'] == 7
aug_filter = irvine_df['Date'].str.startswith('08')

select2_df = irvine_df.loc[flow_filter & aug_filter, :]
select2_df.head()
```

```python
#Groupby
select2_df.groupby(['Date']).max()

select2_df.sort_values('Date')
```
