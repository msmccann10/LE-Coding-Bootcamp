# (1) Delete the unnecessary columns

```python
# Methods would be drop or del. For convenience we will use the drop method. 
bike_df = bike_df.drop(columns=['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'])

bike_df.head()
```

# (2) Rename the Unnamed: 6 column to percent_change.

```python
#Dictionary Method
bike_df = bike_df.rename(columns={'Unnamed: 6': 'percent_change'})

bike_df.head()
```

# (4) Fill all missing values in the 2021 Counts column with 0.

```python
#bike_df.loc[:, '2021 Counts'].fillna(0)
bike_df2 = bike_df
bike_df2[['2021 Counts']] = bike_df2[['2021 Counts']].fillna(0)
bike_df2.head()

# Found this online and like this method. Can speficy multiple columns outputs via dictionary method.
bike_df3 = bike_df.fillna({'2021 Count': 0})
bike_df3.head()
```
