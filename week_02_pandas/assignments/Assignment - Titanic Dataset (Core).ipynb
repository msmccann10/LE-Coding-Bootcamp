This is a dataset from Kaggle which is typically used as an introduction to Kaggle competitions. The usual goal of this dataset is to predict survival on the Titanic based on passenger characteristics.

```python
# Load Data: 
filepath = "/content/drive/MyDrive/Data/titanic.csv" 
cols = ['Name','Survived', 'Pclass', 'Sex', 'Age', 'Ticket', 'Fare', 'Cabin', 'Embarked']
titanic_df = pd.read_csv(filepath, usecols=cols)[cols]

#name_dupes = titanic_df.Name.duplicated()
#titanic_df.loc[titanic_df.Name.duplicated(), :]
```

#(1) Display the head of the dataframe.

```python
display(titanic_df.head())
titanic_df.info()
```

#(2) What is the percentage of people who survived? (hint: find the mean of the survival column).

```python
survival_perc = titanic_df['Survived'].mean()

print(f'The survival rate for the Titanic was {round(survival_perc * 100, 2)}%')
```

#(3) How many of each sex survived?

```python
titanic_df.groupby('Sex')['Survived'].sum()
```

#(4) Considering only the group of passengers who paid less than $10, what percentage of this group survived?

```python
#filter_under10 = titanic_df['Fare'] < 10.0
#fare_under_10 = titanic_df.loc[filter_under10,:]
#under10_survival_rate = fare_under_10['Survived'].mean()
#print(round(under10_survival_rate * 100, 2))

#All in one line...
under10_survival_rate = titanic_df.loc[titanic_df['Fare'] < 10.0,:]['Survived'].aggregate('mean')
#u10 = titanic_df.loc[ titanic_df['Fare']<10, 'Survived'].mean() # THIS IS THE BETTER WAy!!
print(f'The survival rate for passengers who paid less than $10 was {round(under10_survival_rate * 100, 2)}%')
```

#(5) Overall, what is the average age of those who didn't survive?

```python
death_avg_age = titanic_df.loc[titanic_df['Survived'] == False]['Age'].mean()

print(f'The average age of those who died was {round(death_avg_age, 2)}')
```

#(6) Overall, what is the average age of those who did survive?

```python
death_avg_age = titanic_df.loc[titanic_df['Survived'] == True]['Age'].mean()

print(f'The average age of those who survived was {round(death_avg_age, 2)}')
```

#(7) What is the average age of those who did and didn't survive grouped by Sex?

```python
titanic_df.groupby(['Sex', 'Survived'])['Age'].mean().round(2)
```
