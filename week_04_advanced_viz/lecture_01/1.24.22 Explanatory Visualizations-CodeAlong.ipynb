# Explanatory Visualization CodeAlong

- Life Expectancy Data: https://www.kaggle.com/kumarajarshi/life-expectancy-who

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

## Csv file
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRQpH02vlgxAwATTUhJBC0MGiciSz-vUPenWNbrkVH4ijb12NXK-4ut0jLqbANnBgRUo36ZAXDfeQKa/pub?output=csv"
```

```python
df = pd.read_csv(url)
display(df.info())
df
```

# Q1: Does education increase life expectancy?


```python
sns.set_context('talk')
```

```python
ax = sns.scatterplot(data = df, x = 'Schooling', y = 'Life expectancy', hue = 'Status')
#ax.legend(loc=[1.02,0]); -- OR --
ax.legend(bbox_to_anchor=[1,1]);
```

```python
ax = sns.scatterplot(data = df, x = 'Schooling', y = 'Life expectancy', hue = 'Life expectancy', palette = 'Reds')
ax.legend().remove()
ax.set_title("Effect of Schooling on Life Expectancy");
```

> Interpretation

## Q2: Does the development status of a country affect schooling or life expectancy?


```python
fig, axes = plt.subplots(nrows = 2, figsize = (5,10), sharex = True)
sns.violinplot(data = df, y = 'Life expectancy', x = 'Status', ax = axes[0])
sns.violinplot(data = df, y = 'Schooling', x = 'Status', ax = axes[1])
axes[0].set_xlabel('')
axes[1].set_xlabel('')
plt.tight_layout();
```

> Interpretation

# Q3: How have schooling and life expectancy changed over time?

```python
fig, axes = plt.subplots(nrows = 2, sharex = True, figsize = (8,8))
sns.lineplot(data = df, x = "Year", y = "Life expectancy", hue = "Status",  marker = 'o', ci = None, ax = axes[0])
sns.lineplot(data = df, x = "Year", y = "Schooling", hue = "Status", marker = 'o', ci = None, ax = axes[1])
[ax.grid() for ax in axes]
#axes[0].grid() 
#axes[1].grid()
axes[0].legend(bbox_to_anchor=[1,.3]);
axes[1].legend().remove()
plt.tight_layout();
```

> Interpretation
