```python
import json
import pandas as pd
filepath = "D:/Users/Mike PC/Mike PC/.secret/yelp_api.json"

with open(filepath) as f:
    login = json.load(f)
login.keys()
```

```python
from yelpapi import YelpAPI
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
```

```python
results = yelp_api.search_query(location='Philadelphia, PA',
                               term = "Cheesesteak")

type(results)
results.keys()
```
