## Separate a String Column
- separate the display_phone column into area code and phone number

```python
df['display_phone'].head(2)
```

```python
df['display_phone'].str.split(' ', expand=True)
```

```python
df[['area_code', 'phone_number']] = df['display_phone'].str.split(' ', expand=True)
df.drop(columns = ['display_phone', 'phone'], inplace = True)
df.head(2)
```

## Replace Multiple Characters within a String Column
- remove the parentheses from area code

```python
to_replace = ['(',')']

for char in to_replace:
    df['area_code'] = df['area_code'].str.replace(char,'',regex=False)

df['area_code'].head()
```

## Convert String of Dictionaries into Actual Dictionaries
- Coordinates is not actually a dictionary (note the ""... ie it is a string).

```python
coord = df.loc[0,"coordinates"]
print(type(coord))
coord
```

```python
# json.loads(coord)
# fails because of the single quotes areound latitude and longitude
```

```python
coord = coord.replace("'",'"')
coord
```

```python
fixed_coord = json.loads(coord)
print(type(fixed_coord))
fixed_coord
```

```python
df['coordinates'] = df['coordinates'].str.replace("'",'"')
df['coordinates'] = df['coordinates'].apply(json.loads)
df['coordinates'].head()
```

## Unpack columns of dictionaries into separate columns
- Separate our coord columns

```python
lat_long = df['coordinates'].apply(pd.Series)
lat_long
```

```python
df = pd.concat((df, lat_long), axis = 1)
df.drop(columns = ['coordinates'])
df.head(2)
```

## Convert a column containing lists to actual lists
- transactions column is a column containing lists.

```python
df.loc[2,'transactions']
```

```python
df['transactions_split'] = df['transactions'].str.replace("'",'"')
df['transactions_split'] = df['transactions_split'].apply(json.loads)
df['transactions_split'].head()
```

## Convert a column of lists into OHE columns

```python
df['transactions_split'].value_counts()
```

```python
exploded = df.explode('transactions_split')
exploded.head()
```

```python
cols_to_make = exploded['transactions_split'].dropna().unique()
cols_to_make
```

```python
df.head()
```

```python
for col in cols_to_make:
    df[col] = df['transactions'].str.contains(col)

df.head()
```

## Replace String-ids with interger_ids
replace long string ids with a numeric version 

```python
unique_ids = sorted(df['id'].unique())
unique_ids[:3]
```

```python
int_ids = range(len(unique_ids))
int_ids
```

```python
# Zip together into a dictionary
id_map = dict(zip(unique_ids, int_ids))
id_map
```

```python
# Apply map to df to get int_ids
df['id'] = df['id'].map(id_map)
df.head()
```

```python
# Create and id lookup table for SQL
id_lookup = pd.DataFrame({'str_id':id_map.keys(),
                         'int_id':id_map.values()})
id_lookup.head(3)
```
