#ETL
Extract, Transform, Load... data from mulitple sources to a data warehouse

## Extract
Data is copied or exported from locatiosn to a staging area

includes but not limited to. Raw/flat files, APIs, Emails, Websites, CRM systems, SWL databases, etc.

## Transform
Once in a central location the data is transformed and consolidated for its intended analytic use. This can include:
- Filtering, cleaning, de=duplicating, validating, etc
- Performing calculations, translations, summarizations
- Removing, encrypting, or protecting data 
- formating into tables or joined tables for target warehouse or relational database
- audinting data to ensure quality and compliance

## Load
Once transformed the data is moved from the staging area into a target data warehouse or database.

# Working with JSON
JavaScript Object Notation

JSON files are text-based files structured into dictionaries or list of dictionaries (also called records). JSON is used to transfer data between computers and programs. Considered to be 'language independent'.

## Open Function
open creates a variable that stays connected to the opened file.
- requires a filepath 
- and a mode ("r" for read text, "w' for write text, "rb" for read binary, "wb" for write binary)

BEST PRACTICE: use the with command to open a temporary file to avoid corrupting the file/path.

```python
message = """My test message to save to file.
it is a multi-string message"""
message
```

```python
with open ('example_file.txt','w') as file:
    file.write(message)
```

```python
with open('example_file.txt','r') as f:
    loaded = f.read()
loaded
```

## JSON Functions
4 basic functions... 2 we will use regularly for saving and loading data:
- json.load(file): Loads the data from a json file into a variable (dictionary of list of dictionaries).
- json.dump(data, file): Saves the JSON-compatible data into a file.

2 we will use less... for converting between a string and a JSON directory:
- json.loads: Convert an already-loaded string into a dictionary or list of dictionaries
- json.dumps: Convert a dictionary or list of dictionaries into a string.

# Concepts of Efficient API Extraction

## API Limits
- APIs usually limit the nubmer of API calls that you can perfomr in a specific period (day, week, month).
- Free tier typically limits the number, premium teirs unlock more...
- Check the API before running it to understand the number of times you can run it before you are out.

## Why Efficient API
Possible problems that could eat up your API limit...
- Interrupted workflow/kernal
- Unexpected Errors
- Frozen Loop
- Overwriting Previous Results
- Duplicating API Calls

Putting safeguards in place can avoid many (all?) of these problems and save your API limit (and time). 
- Create an empty results file ONLY if it doesnt already exist
- Save your API results within your loop so you have every API call saved
- Add a progress bar to your loop so you can tell if its frozen vs running.
- Double check your list of API calls against you previously saved results.
