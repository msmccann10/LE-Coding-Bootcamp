# Boosting
Ensamble modeling techniques are meant to combine the results of multiple models to overcome the shortcomings of a single model. Examples include:
- Bagging
- Boosting
- Stacking

Boosting: Sequentialy use the results of weak learners to improve the performance of the final model. Used to reduce model bias.

Gradient Boosting: Sequentially fitting models on the erors of the previous models. 

## Gradient Boosting

Sequentially fit model to the errors of the previous models. 
- Initialize a constant baseline prediction
- compute the residual errors of the value of the loss function using prediction from the previous model. 
- construct a decision tree using the features to predict the residuals 
- add the prediction from the previous model to the new prediction (times learning rate)
- repeat. 


### Python Example

```python
# from sklearn.ensemble import GradientBoostingClassifier
```

```python
df = pd.read_csv(cancer_filepath, index_col = 'id')
df.head()
```

```python
df.drop(columns = 'Unnamed: 32', inplace = True)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
X = df.drop(columns = 'diagnosis')
y = df['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 3)
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(X_train, y_train)
```

```python
print('training accuracy:', gbc.score(X_train, y_train))
print('test accuracy:', gbc.score(X_test, y_test))
```

## LightGBM
LightGBM is a library that allows us to create efficient gradient boosting machines. It adds onto a classic gradient boosting algorithm by adding an automated feature selection, which ultimately speeds up training and improves performance.
- Builds leaf wise


### Python Example

```python
#from lightgbm import LGBMClassifier
```

```python
df = pd.read_csv(cancer_filepath, index_col = 'id')
df.head()
```

```python
df.drop(columns = 'Unnamed: 32', inplace = True)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 2})
```

```python
X = df.drop(columns = 'diagnosis')
y = df['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 3)
```

```python
lgbm = LGBMClassifier()
lgbm.fit(X_train, y_train)
```

```python
print('Training Accuracy:', lgbm.score(X_train, y_train))
print('Test Accuracy:', lgbm.score(X_test, y_test))
```

## XGBoost
XGBoost, which stands for eXtreme Gradient Boosting, is an implementation of the Gradient Boosting algorithm that is designed for speed, efficient memory, and high performance.
- Builds level/depth wise

### Python Example

```python
# from xgboost import XGBClassifier
```

```python
df = pd.read_csv(cancer_filepath, index_col = 'id')
df.head()
```

```python
df.drop(columns = 'Unnamed: 32', inplace = True)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 2})
```

```python
X = df.drop(columns = 'diagnosis')
y = df['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 3)
```
