```python
class User:
    def __init__(self, username, email_address):
        self.name = username
        self.email = email_address
        self.account = BankAccount(int_rate=0.02, balance=0)

class BankAccount:
    def __init__(self, int_rate, balance):
        self.interest = int_rate
        self.balance = balance
    def deposit(self, ammount):
        self.balance += ammount
        print(f"\nDeposit complete. After your {ammount} deposit, you now have {self.balance} in your account. Have a nice day!")
    def withdrawal(self, ammount):
        self.balance -= ammount
        print(f"\nWithdrawal complete! After your {ammount} withdrawal, you now have {self.balance} in your account. Have a nice day!")
    def transfer(self, ammount, tran_user):
        self.balance -= ammount
        tran_user.account.balance += ammount
        print((f"\nTransfer complete! You have transfered {ammount} to {tran_user.name}"))


guido = User("Guido van Rossum", "guido@python.com")
monty = User("Monty Python", "monty@python.com")

guido.account.deposit(150)
guido.account.withdrawal(25)
guido.account.deposit(75)
guido.account.transfer(100, monty)

print(guido.account.balance)
print(monty.account.balance)
```
