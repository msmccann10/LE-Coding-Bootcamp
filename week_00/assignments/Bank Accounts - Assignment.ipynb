```python
class BankAccount:
    def __init__(self, int_rate = 0.01, balance = 0):
        self.interest = int_rate
        self.balance = balance
    def deposit(self, ammount):
        self.balance += ammount
        print(f"\nDeposit complete! After your {ammount} deposit, you now have {self.balance} in your account. Have a nice day!")
    def withdrawal(self, ammount):
        if(self.balance - ammount >= 0):
            self.balance -= ammount
            print(f"\nWithdrawal complete! After your {ammount} withdrawal, you now have {self.balance} in your account. Have a nice day!")
        else:
            print("\ninsufficient funds: Charging a $5 fee")
            self.balance = self.balance - 5.00
    def transfer(self, ammount, tran_user):
        self.balance -= ammount
        tran_user.account.balance += ammount
        print((f"\nTransfer complete! You have transfered {ammount} to {tran_user.name}"))
    def display_account_info(self):
        print((f"\nAccount balance is {self.balance}. Have a nice day!"))
    def yield_interest(self):
        self.balance = self.balance + (self.balance * self.interest)
```

```python
guidoaccount = BankAccount()
montyaccount = BankAccount()
```

```python
guidoaccount.deposit(100)
guidoaccount.deposit(50)
guidoaccount.deposit(250)
guidoaccount.withdrawal(20)
guidoaccount.yield_interest()

guidoaccount.display_account_info()

#test if/else for insufficeint funds
#guidoaccount.withdrawal(400)
#guidoaccount.display_account_info()
```
