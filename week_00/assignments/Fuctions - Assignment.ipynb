(1) **Countdown** - Create a function that accepts a number as an input. Return a new list that counts down by one, from the number (as the 0th element) down to 0 (as the last element).
>Example: countdown(5) should return the list: [5,4,3,2,1,0]

```python
def countdown(x):
    list = []
    for i in range(x, 0, -1):
        list.append(i) 
    return list

countdown(5)
```

```python
def countdown2(y):
    list2 = []
    while y >= 0:
        list2.append(y)
        y -= 1
    return list2

countdown2(5)
```

(3) **First Plus Length (Optional)** - Create a function that accepts a list and returns the sum of the first value in the list plus the list's length.
>Example: first_plus_length([1,2,3,4,5]) should return 6 (first value: 1 + length: 5)

```python
def first_plus_last(x):
    fpl = x[0] + len(x)
    return fpl

first_plus_last([1,2,3,4,5])
```

(4) **This Length, That Value (Optional)** - Write a function that accepts two integers as parameters: size and value. The function should create and return a list whose length is equal to the given size, and whose values are all the given value.
>Example: length_and_value(4,7) should return [7,7,7,7]<br>
>Example: length_and_value(6,2) should return [2,2,2,2,2,2]

```python
def length_and_value(x,y):
    list = []
    while x > 0:
        list.append(y)
        x -= 1
    return list

list1 = length_and_value(4,7)
list2 = length_and_value(6,2)

print(list1)
print(list2)
```
