```python
# Counting, the Dojo Way (Optional) - Print integers 1 to 100. If divisible by 5, print "Coding" instead. 
#If divisible by 10, print "Coding Dojo".
for i in range(101):
    if(i % 10 == 0):
        print("Coding Dojo")
    elif(i % 5 ==0):
        print("Coding")
    else:
        print(i)
```

```python
# Whoa. That Sucker's Huge(Optional) - Add odd integers from 0 to 500,000, and print the final sum.
sum = 0
for i in range(500000):
    if(i % 2 != 0):
        sum += i

print("The sum of all the odd values between 0 and 500,000 is ", sum, "!", sep = "")

#alternatively
my_list = []
total = 0

for i in range(500000):
    if(i % 2 != 0):
        my_list.append(i)

for i in range(0, len(my_list)):
    total += my_list[i]

print("Or we can do it this way too! The answer is ", total, "!", sep = "")
```
